```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("C://Users//shrad//OneDrive//Desktop//Python Practice jupyter notebook//UNZIP_FOR_NOTEBOOKS_FINAL//05-Seaborn//dm_office_sales.csv")
```

```python

plt.figure(figsize=(12,4),dpi=200)
#hue asks for another column based on the categories and create a legend 
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')

plt.show()
```

```python

plt.figure(figsize=(12,4),dpi=200)
#hue asks for another column based on the categories and create a legend 
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')

plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200,edgecolor='blue')
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',s=10)
plt.show()
plt.savefig('Scatterplot')
```

```python
# distribution plots display a single continous feature and help visualize properties such as std and mean
# three main distribution plot type
# rug plot, histogram, and KDE Plot (kernel distribution estimation)

#rugplot

sns.rugplot(x='salary',data=df)
```

```python
df.head()
```

```python
sns.rugplot('salary',data=df,height=0.5)
```

```python
#do not use DISTPLOT 

#distribution plot
sns.set(style='darkgrid')
#sets the style for everything in the notebook
sns.displot(data=df,x='salary',bins=20,color='brown',kde=True,rug=True) #to get kde and rug choose kde=True, rug=True
plt.show()
```

```python
sns.histplot(x='salary',data=df)
plt.show()
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.set_title('The KDE plot for Salary')
axes.set_xlabel('density dist')
axes.set_ylabel('Saalry values')
sns.kdeplot(x='salary', data=df)

plt.show()
```

```python
sns.kdeplot(x='age',data=sample_ages,bw_adjust=0.5,clip=[0,100],shade=True) #shade=True shades under area
#clip clips off the kde, bw_adjust if very small, really thin normal dist, as bw_adjust increases, the shape looks more like normal
```

```python
df=pd.read_csv("C://Users//shrad//OneDrive//Desktop//Python Practice jupyter notebook//UNZIP_FOR_NOTEBOOKS_FINAL//05-Seaborn//StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='lunch')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(data=df,y='reading score',x='parental level of education')
```

```python
plt.figure(figsize=(12,5),dpi=300)
sns.swarmplot(data=df,y='reading score',x='parental level of education',hue='lunch')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.boxenplot(data=df,y='reading score',x='parental level of education',hue='lunch')
```

```python
#comparison plots are 2D versions of the plots
#CPU RAM intensive for large dataframes
#pairplot() #to check comparison
plt.figure(figsize=(12,5),dpi=300)
sns.jointplot(data=df,x='reading score',y='math score',hue='lunch',kind='kde',shade=True)
```

```python
sns.pairplot(df,hue='lunch')
```

```python
#seaborn grids
#calls use matplotlib sublots() to automatically create a grid based off a categorical column
#instead of passing in a specific number of cols and rows for the subplots, we can pass the name of the cols and seaborn automatically map the subplot grids
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch') #plot for every row (lunch)
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch') #plot for every col (lunch)
```

```python
#gridplots

g=sns.PairGrid(df) #create a grid

g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.displot)
g=g.map_lower(sns.histplot)
```
